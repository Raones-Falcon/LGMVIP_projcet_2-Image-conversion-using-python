```python

# OpenCV library for image processing.
# We'll load the image using the imread() function from OpenCV.
# 

import cv2
import matplotlib.pyplot as plt
```

```python
cv2.imshow("Uploaded image ",image)     # read and show the image  in new tab
cv2.waitKey(0)
cv2.destroyAllWindows()
```
