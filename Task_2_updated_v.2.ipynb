```python
#pip.install opencv-python


# 1. import


import cv2
import matplotlib.pyplot as plt
```

```python
# Read image 

pic=cv2.imread('AA.jpg')
```

```python
# cretaing a  new window  to show uploded image :




cv2.imshow('Uploded image',pic)
cv2.waitKey(5)
```

```python
# Step-1 : GrayScale image conversion

gray_img=cv2.cvtColor(pic,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray',gray_img)
cv2.waitKey(6)
```

```python
# Step-2 : inverting gray image 


inverted_image=255-gray_img
cv2.imshow('Inverted gray image',gray_img)
cv2.waitKey(7)
```

```python
# step-3 : blurr the inverted image

blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
```

```python
# step-5 : inverted blurr

inverted_blurred= 255-blurred
pencil_sketch=cv2.divide(gray_img,inverted_blurred,scale=256.0)
cv2.imshow('sketch',pencil_sketch)
cv2.waitKey(0)
```
